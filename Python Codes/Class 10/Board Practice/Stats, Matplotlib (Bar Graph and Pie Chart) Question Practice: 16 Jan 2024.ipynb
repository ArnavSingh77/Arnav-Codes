```python
'''
Heights of 25 children (in cm) in a school are as given below:
168, 165, 163, 160, 163, 161, 162, 164, 163, 162, 164, 163, 160, 163, 164, 163, 160, 165, 163, 162
Write python code to find mean, median and mode of heights.
'''
import statistics as stat

height = [168, 165, 163, 160, 163, 161, 162, 164, 163, 162, 164, 163, 160, 163, 164, 163, 160, 165, 163, 162]
print("The Mean of the data is: %.2f"% stat.mean(height))
print("The Median of the data is: %.2f"% stat.median(height))
print("The Mode of the data is: %.2f"% stat.mode(height))
print("The Standard Deviation of the data is: %.2f"% stat.stdev(height))
print("The Variance of the data is: %.2f"% stat.variance(height))
```

```python
'''
Following data gives total marks (out of 600) obtained by six children of a particular class. Write PYTHON code to represent the data on a bar graph.
'''
import matplotlib.pyplot as plt
students = ['Ajay', 'Bali', 'Dipti', 'Faiyaz', 'Geetika', 'Hari','Arnav']
marks = [450, 500, 300, 360, 400, 540, 600]

bar = plt.bar(students,marks, color=['red','green','blue','magenta','purple','orange','black'])
plt.title("Student Marks")
plt.xlabel("Name")
plt.ylabel("Marks Scored")
#plt.legend(bar, students)
plt.grid(True)
plt.show()
```

```python
'''
The number of Mathematics books sold by a shopkeeper on six consecutive days is shown below:
WAP in python using Matplotlib to draw a bar chart to represent the above information.
'''
import matplotlib.pyplot as plt

day=['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday']
books=[65,40,30,50,20,70]

bar = plt.bar(day,books,color=['blue','black','green','red','yellow','orange'])
plt.xlabel("Days")
plt.ylabel("Number of Books Sold")
plt.legend(bar,day)
plt.grid(True)
plt.show()
```

```python
'''
On a particular day, the sales (in rupees) of different items of a bakers shop are given below:
WAP in python using Matplotlib to draw a pie chart for this data.
'''

import matplotlib.pyplot as plt

item=['Ordinary Bread','Fruit Bread','Cakes & Pasteries','Biscuits','Others']
price=[320,80,160,120,40]

plt.pie(price, labels=item)
plt.legend(loc='lower right')
plt.show()
```
